```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')

df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})


```

```python
df.columns
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)


```

```python
df[['age', 'age_group']].head(10)
```

```python
#create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# Display the first 10 rows of the specified columns
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
from sqlalchemy import create_engine
import psycopg2

# Step 1: PostgreSQL details
username = "postgres"          # your postgres user
password = "1234"              # <-- put your real password here
host = "localhost"             # keep as localhost
port = "5432"                  # default PostgreSQL port
database = "customer_behavior" # database you created in pgAdmin

# Step 2: First connect to default 'postgres' database to create our database if it doesn't exist
default_conn = psycopg2.connect(
    user=username,
    password=password,
    host=host,
    port=port,
    database="postgres"  # Connect to default database first
)
default_conn.autocommit = True
cursor = default_conn.cursor()

# Check if our database exists, if not create it
cursor.execute(f"SELECT 1 FROM pg_catalog.pg_database WHERE datname = '{database}'")
exists = cursor.fetchone()
if not exists:
    print(f"Creating database '{database}'...")
    cursor.execute(f"CREATE DATABASE {database}")
    print(f"Database '{database}' created successfully.")
else:
    print(f"Database '{database}' already exists.")

cursor.close()
default_conn.close()

# Step 3: Now create engine to connect to our database
engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

# Step 4: Load DataFrame into PostgreSQL
table_name = "customer"        # table name in PostgreSQL

# Make sure df is defined before this line
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
